```python
import numpy as np
import pandas as pd
```

```python
Country_data=pd.read_csv("/content/3.01. Country clusters.csv")
```

```python
Country_data.head(15)
```

```python
New_Country_data=Country_data.iloc[:,1:3]
```

```python
New_Country_data.head()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3) #K=3
```

```python
kmeans.fit(New_Country_data)
```

```python
clusters=kmeans.fit_predict(New_Country_data)
```

```python
print(clusters)
```
